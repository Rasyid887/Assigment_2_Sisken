```python
import numpy as np
import matplotlib.pyplot as plt
import control as co
```

```python
n = 150
T = 1
persamaan01 = co.tf([1],[T,(T/16),1])
persamaan01
```

```python
#Input step
t = np.linspace(0,n,1000)
t1, y1 = co.step_response(persamaan01, t)

plt.figure()
plt.plot(t1, y1, label = 'Input step')
plt.grid()
plt.title('Input Step Response')
plt.xlabel('Time')
plt.ylabel('C(t)')
plt.legend()
plt.show()
```

```python
#Input ramp
n = 30
t = np.linspace(0,n,1000)
u = t
t2, y2 = co.forced_response(persamaan01, t, u)

plt.figure()
plt.plot(t, u, label='input signal')
plt.plot(t2, y2, label='ramp response')
plt.grid()
plt.title('Input Ramp Response')
plt.xlabel('Time')
plt.ylabel('C(t)')
plt.legend()
plt.show()
```

```python
#Input impulse
t3, y3 = co.impulse_response(persamaan01, t)

plt.figure()
plt.plot(t3, y3, label = 'Input impulse')
plt.grid()
plt.title('Input impulse Response')
plt.xlabel('Time')
plt.ylabel('C(t)')
plt.legend()
plt.show()
```
